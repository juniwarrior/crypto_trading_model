```python
#Preparamos las variables para iterar
bs = BinanceSocketManager(client)
socket = bs.trade_socket('BTCUSDT')
```

```python
#Con esta funcion se pondran nuestros datos en un DataFrame estructurado
def createframe(msg):
  df = pd.DataFrame([msg])
  df = df.loc[:,['s','E','p']]
  df.columns = ['symbol', 'Time', 'Price']
  df.Price = df.Price.astype(float)
  df.Time = pd.to_datetime(df.Time, unit = 'ms')
  return df
```

```python
# Creador: https://www.youtube.com/watch?v=rc_Y6rdBqXM
#La funcion itera el precio de BTC hasta x = 10
#No sabemos que es aenter
x = 0
while x != 10:
  await socket.__aenter__()
  msg = await socket.recv()
  frame = createframe(msg)
  x = x + 1
  print(frame)
```
